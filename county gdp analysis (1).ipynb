## intro to seaborn

```python
#seaborn is very powerfull visualization(STATISTIC VISUALIZATION) pkg in python
import matplotlib.pyplot as plt # visualization
import seaborn as sns # distribution visualization

%matplotlib inline
plt.rcParams['figure.figsize']=8,4

#import warnings
#warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
#distributions:
vis1=sns.displot(stats["InternetUsers"],kde=True)
```

```python
vis1=sns.displot(stats["InternetUsers"],kde=False)
```

```python
vis1=sns.displot(stats["InternetUsers"],kde=True,bins=10)
```

## boxplots

```python
vis2=sns.boxplot(data=stats,x="IncomeGroup",y="BirthRate")
```

```python
vis2=sns.boxplot(data=stats,x="IncomeGroup",y="BirthRate",palette="Set2",hue=None)
plt.show()
```

```python
# visualizing with seaborn
```

```python
vis3=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False) #lm=linear model
```

```python
vis4=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate')
```

```python
vis4=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup') #hue=parameter
```

```python
vis5=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup',height=5) #height defines the height of the graph
plt.show()
```
